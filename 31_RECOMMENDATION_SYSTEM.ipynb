```python
#change the structure of dataset to compute the similarity score
movie_df=movie.pivot(index='userId', #row=userId
                     columns='movie', #columns=movie
                     values='rating')
```

```python
movie_df
#NaN means person didn't watch the movie so he didnt gived ratings.so now we can recommend movie that he didn't watch
```

```python
#impute those NaN values with 0,didn't watch movie
movie_df.fillna(0,inplace=True)
movie_df
```

```python
#calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
#user similarity-pairwise-taking 2 rows at a time it will compute distance
#the distance is  scaled between 0 and 1
#suppose the distance between 2 rows is 0.we want similarity not a distance
#so,now it is calculated by similarity=1-distance,so 1-0=1, i.e., similaity between 2 rows is 1
#suppose distance is 0.9 now 1-0.9=0.1
user_sin=1-pairwise_distances(movie_df.values,metric='cosine')
```

```python
user_sin#This is similarity matrix in which diagnol is having 1 value
```

```python
#Nullifying the diagonal values
np.fill_diagonal(user_sin,0)
```

```python
#store the results in dataframe
user_sin_df=pd.DataFrame(user_sin)
user_sin_df
```

```python
#set index and column names to user id
user_sin_df.index=movie.userId.unique()
user_sin_df.columns=movie.userId.unique()
```

```python
user_sin_df
```

```python
#idxmax() method returns a series of index with the maximum  value for each column(row 3 and column 11 has the highest similarity)
#by speciifying the column axis(axis=1)the idxmax will return a series of indexof the maxium value
user_sin_df.idxmax(axis=1)
```

```python
#most similar values(top 10)
user_sin_df.idxmax(axis=1)[0:10]
```

```python
#5x5 matrix
user_sin_df.iloc[0:5,0:5]
#similarity 1-between customer 3 to 3 and 6 to 6
#similarity between customer 3 and 11 is 1.i.e., they are very similar
```
