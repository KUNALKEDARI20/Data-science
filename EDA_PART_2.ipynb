**ISOLATION FOREST FOR OUTLIER DETECTION**

---




```python
from sklearn.ensemble import IsolationForest
import pandas as pd
```

```python
df=pd.DataFrame({
    'Marks':[50,80,82,95,800],
    'Time':[1,2,3,1.5,60]
})
```

```python
df
```

```python
# What is outlier here
#iso=IsolationForest(random_state=10),excepting 10% outlier in dataset

iso=IsolationForest(random_state=10,contamination=0.1)
iso.fit(df)

out=iso.predict(df)
```

```python
out # -1: last value is outlier
```

```python
df['Score']=out
df
```

```python
#lets workout isolation forest on iris dataset

data=pd.read_csv('/content/Iris.csv',index_col=0)
data_encoded=pd.get_dummies(data)
data_encoded
```

```python
# u can detect outliers using percentiles as well.other than this boxplot you can use for outlier detection
data_encoded['PetalWidthCm'].describe(percentiles=[0.1,0.02,0.9,0.99,0.991,0.992])
```

```python
# training the model
clf=IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded)
#contamination=how much percentage of outliers you are expecting in dataset
#eg.in health care domain it will be very low 1% or less(0.01 or 0.001)
```

```python
#predictions

pred_outliers=clf.predict(data_encoded)
pred_outliers # -1 are outliers and 1 are inliners
```

```python
data_encoded.shape
```

```python
# purposly let us add new data point to the dataset
data_encoded.loc[151]=[20,40,30,50,1,0,0]
data_encoded.shape
```

```python
# training the model
clf=IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded)

# Predictions
y_pred_outliers=clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded['anomaly']=clf.predict(data_encoded.iloc[:,0:7])
```

```python
# Syntax:pps.score(df,'feature_column','target_column')
pps.score(data,'SepalLengthCm','PetalLengthCm') #ppscore=0.55 it is ok kind of score
```

```python
#calculate the whole pps matrix
pps.matrix(data) # petal width with species is most important variable with pps score with 0.9276
```

```python
df=pd.DataFrame(pps.matrix(data))
df[df.case!='predict_itself'].iloc[:,[0,1,2,3,5,8]].sort_values('ppscore',ascending=False)
#PPSCORE is higher in petalwidthcm and species i.e0.927652 petal width is very good feature in predicting species
```
