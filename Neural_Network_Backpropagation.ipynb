```python
# Create your first MLP in Keras. Keras is a library built on top of Tensar flow. It is used to build NN as well as to optimize hyperparameter.
from keras.models import Sequential # In NN we are going to add components (i.e. hidden layers) one by one in sequential manner
from keras.layers import Dense
# Keras Dense layer is the layer that contains all the neurons that are deeply connected within themselves.
#This means that every neuron in the dense layer takes the input from all the other neurons of the previous layer.
import numpy as np
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
# load pima indians dataset
dataset = np.loadtxt("/content/pima-indians-diabetes.data.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
dataset.shape
```

```python
#create model
model=Sequential()
model.add(Dense(20,input_dim=8,activation='relu'))#1st layer
model.add(Dense(10,activation='relu'))#2nd layer
model.add(Dense(10,activation='relu'))#3rd layer
model.add(Dense(1,activation='sigmoid'))#output layer
```

```python
#Compile model-Establish connection between input layer,hidden layer,output layer
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
# binary_crossentropy - Computes the crossentropy metric between the labels and predictions.
# This is the crossentropy metric class to be used when there are only two label classes (0 and 1)
```

```python
#fit the model
history=model.fit(X,Y,batch_size=10,epochs=250,validation_split=0.33)
```

```python
# summarize history for accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
