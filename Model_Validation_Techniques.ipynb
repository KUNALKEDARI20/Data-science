**MODEL VALIDATION TECHNIQUES**

---




```python
# Import Libraries
import pandas as pd
from pandas import read_csv
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Load Dataset
filename='/content/pima-indians-diabetes.data.csv'
cols=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(filename,names=cols)

array=dataframe.values
```

```python
dataframe
```



---
**1.** **Train-Test Split**

---




```python
# Split the data into X and Y
X=array[:,0:8]
Y=array[:,8]
```

```python
# Building Train-Test split

test_size=0.33
seed=7 # random_state=7(seed)

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=test_size,random_state=seed)
model=LogisticRegression() # Train the model for Logistic regression as our Y is categorical variable(yes\no) variable
model.fit(X_train,Y_train)

result=model.score(X_test,Y_test)
result
```

```python
# if you want to convert the result into percentgae
result*100
```

**2.K-Fold Cross Validation**

---




```python
# Load Libraries
import pandas as pd
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
# Load Dataset
filename='/content/pima-indians-diabetes.data.csv'
cols=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(filename,names=cols)
```

```python
dataframe
```

```python
array=dataframe.values
X=array[:,0:8]
Y=array[:,8]
```

```python
num_folds=10

kfold=KFold(n_splits=num_folds)
model=LogisticRegression(max_iter=200)
results=cross_val_score(model,X,Y,cv=kfold)
```

```python
results # accuracy of mean model
```

```python
results.mean()*100 # final accuracy is mean of all accuracies
```

**3.Leave One Out Cross Validation Technique**

---




```python
import pandas as pd
from pandas import read_csv
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
# Load Dataset
filename='/content/pima-indians-diabetes.data.csv'
cols=['preg','plas','pres','skin','test','mass','pedi','age','class']
dataframe=read_csv(filename,names=cols)

array=dataframe.values
X=array[:,0:8]
Y=array[:,8]
```

```python
# LOOCV
loocv=LeaveOneOut()
model=LogisticRegression(max_iter=300)
results=cross_val_score(model,X,Y,cv=loocv)
```

```python
results # 1 indicates 100% accuracy and 0 indicates 0% accuracy here
```

```python
results.mean()*100
```
